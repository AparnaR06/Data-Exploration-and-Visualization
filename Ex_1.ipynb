```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create sample data
data = {
    'Product': ['Laptop', 'Tablet', 'Smartphone', 'Monitor', 'Keyboard'],
    'Units Sold': [120, 80, 300, 150, 200],
    'Price': [70000, 25000, 30000, 15000, 2000]
}

# Step 2: Create a DataFrame
df = pd.DataFrame(data)

# Step 3: Display the DataFrame
print("Sales Data:\n")
print(df)

# Step 4: Data Analysis
print("\n--- Data Analysis ---")
print("Total Revenue per Product:")
df['Revenue'] = df['Units Sold'] * df['Price']
print(df[['Product', 'Revenue']])

print("\nAverage Units Sold:", df['Units Sold'].mean())
print("Highest Selling Product:", df.loc[df['Units Sold'].idxmax(), 'Product'])

# Step 5: Data Visualization
plt.figure(figsize=(10, 5))
plt.bar(df['Product'], df['Units Sold'], color='skyblue')
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.title('Units Sold per Product')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
