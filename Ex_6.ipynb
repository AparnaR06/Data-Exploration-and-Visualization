```python
# Install required libraries
!pip install folium pandas

# Import libraries
import pandas as pd
import numpy as np
import folium
from folium.plugins import MarkerCluster

# STEP 1: Generate random data
np.random.seed(0)
latitudes = np.random.uniform(low=37.5, high=38.5, size=100)
longitudes = np.random.uniform(low=-123, high=-121, size=100)
values = np.random.randint(1, 100, size=100)

# STEP 2: Create DataFrame
df = pd.DataFrame({
    'LATITUDE': latitudes,
    'LONGITUDE': longitudes,
    'VALUE': values
})

# STEP 3: Display Basic Statistics
print("OUTPUT")
print("\nBasic Statistics:\n")
print(df.describe())

# STEP 4: Create Interactive Map
base_map = folium.Map(location=[df['LATITUDE'].mean(), df['LONGITUDE'].mean()], zoom_start=8)
marker_cluster = MarkerCluster().add_to(base_map)

# STEP 5: Add Points with Tooltip
for index, row in df.iterrows():
    tooltip_text = f"Value: {row['VALUE']}"
    folium.Marker(
        location=[row['LATITUDE'], row['LONGITUDE']],
        tooltip=tooltip_text
    ).add_to(marker_cluster)

# STEP 6: Display the Map
base_map
```
