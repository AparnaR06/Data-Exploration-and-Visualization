```python
import numpy as np

# Create a 3x3 array with random integers
arr = np.random.randint(1, 10, size=(3, 3))
print("Generated Array:\n", arr)

# Array properties
print("Shape of array:", arr.shape)
print("Size of array:", arr.size)
print("Number of dimensions:", arr.ndim)
print("Data type:", arr.dtype)

# Slicing - Get second row
print("Second row of the array:", arr[1])

# Slicing - Get first column
print("First column of the array:", arr[:, 0])
```

```python
import pandas as pd
import numpy as np

# Create DataFrame from dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 22],
    'Marks': [85, 90, 95]
}

df = pd.DataFrame(data)
print("DataFrame:\n", df)

# Describe the DataFrame
print("\nSummary Statistics:\n", df.describe())

# Create DataFrame from 2D NumPy array
array_data = np.array([[100, 200], [300, 400]])
df2 = pd.DataFrame(array_data, columns=['A', 'B'])
print("\nDataFrame from 2D Array:\n", df2)

# Access specific column
print("\nAge column:\n", df['Age'])
```

```python
import matplotlib.pyplot as plt

# Line plot
x = np.arange(1, 6)
y = np.array([2, 4, 6, 8, 10])
plt.plot(x, y, marker='o', color='green')
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()

# Bar chart
students = ['A', 'B', 'C', 'D']
marks = [75, 85, 90, 95]
plt.bar(students, marks, color='orange')
plt.title("Students Marks")
plt.ylabel("Marks")
plt.show()

# Subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# First subplot
axes[0].plot(x, y, 'b-*')
axes[0].set_title("Line Plot")

# Second subplot
axes[1].bar(students, marks, color='purple')
axes[1].set_title("Bar Chart")

plt.tight_layout()
plt.show()
```
