```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Create a Unique E-Commerce Order Dataset
data = {
    'OrderID': range(1001, 1011),
    'Customer': ['Anita', 'Ravi', 'Sunil', 'Megha', 'Priya', 'Ravi', 'John', 'Megha', 'Sunil', 'Anita'],
    'Product': ['Laptop', 'Mobile', 'Headphones', 'Laptop', 'Tablet', 'Mobile', 'Smartwatch', 'Tablet', 'Laptop', 'Smartwatch'],
    'Price': [55000, 20000, 3000, 58000, 15000, 22000, 8000, 17000, 60000, 7500],
    'Quantity': np.random.randint(1, 5, size=10),
    'OrderDate': pd.date_range(start='2024-01-01', periods=10, freq='D')
}

df = pd.DataFrame(data)

# Step 3: Explore the Dataset
print("First 5 rows:\n", df.head())
print("\nSummary Statistics:\n", df.describe(include='all'))
print("\nMissing Values:\n", df.isnull().sum())
print("\nUnique Products:\n", df['Product'].unique())
print("\nUnique Customers:\n", df['Customer'].unique())

# Step 4: Data Cleaning
df.drop_duplicates(inplace=True)

# Step 5: Apply Variable Filters
selected_columns = df[['Customer', 'Product', 'Price']]
print("\nSelected Columns:\n", selected_columns)

# Step 6: Apply Row Filters
# Orders with Price greater than 20,000
high_value_orders = df[df['Price'] > 20000]
print("\nHigh Value Orders (Price > 20,000):\n", high_value_orders)

# Orders of Laptops
laptop_orders = df[df['Product'] == 'Laptop']
print("\nLaptop Orders:\n", laptop_orders)

# Step 7: Data Visualization

# Price Distribution
plt.hist(df['Price'], bins=5, edgecolor='black', color='lightgreen')
plt.title('Product Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Quantity Boxplot
plt.boxplot(df['Quantity'])
plt.title('Quantity Distribution')
plt.ylabel('Quantity')
plt.show()

# Product-wise Order Count
product_counts = df['Product'].value_counts()
product_counts.plot(kind='bar', color='skyblue')
plt.title('Product-wise Order Count')
plt.xlabel('Product')
plt.ylabel('Count')
plt.show()

# Pie Chart for Customer Orders
customer_counts = df['Customer'].value_counts()
customer_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Customer Order Share')
plt.ylabel('')
plt.show()
```
